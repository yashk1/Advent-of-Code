```python
f = open("/Users/yash/github/Advent of code/input/day3.txt",'r+')
l = f.read().splitlines()

l
```

```python
sum = 0
for line in l:
    rs1, rs2 = line[0: int(len(line)/2)], line[int(len(line)/2):]
    common = [match for match in rs1 if match in rs2][0]
    if common.isupper():
        sum+= ord(common) - 38
    else:
        sum += ord(common)-96

print(sum)
    
```
